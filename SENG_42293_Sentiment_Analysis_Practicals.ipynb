```
import pandas as pd
```

```
movie_review_data = pd.read_csv("user_movie_review.csv")
movie_review_data.shape
```

```
movie_review_data.head()
```

```
movie_review_data['class'].value_counts()
```

```
from sklearn.feature_extraction.text import CountVectorizer
```

```
countervectemp = CountVectorizer()
dtm_temp = pd.DataFrame(countervectemp.fit_transform(movie_review_data['text']).toarray(),
                        columns=countervectemp.get_feature_names(), index=None)
dtm_temp['class'] = movie_review_data['class']
dtm_temp.head()
```

```
dtm_temp['class'][0:3]
```

```
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
import re
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
```

```
stemmer = PorterStemmer()

def tokenize(text):
  text = stemmer.stem(text)
  text = re.sub(r'\W+|\d+|_', ' ', text)
  tokens = nltk.word_tokenize(text)
  return tokens
```

```
countervect = CountVectorizer(min_df=5, tokenizer=tokenize, stop_words=stopwords.words('english'))
dtm = pd.DataFrame(countervect.fit_transform(movie_review_data['text']).toarray(),
                        columns=countervect.get_feature_names(), index=None)
dtm['class'] = movie_review_data['class']
dtm.head()
```

```
mv_train = dtm[:1900]
mv_test = dtm[1900:]
```

```
from sklearn.naive_bayes import MultinomialNB
```

```
x_train = mv_train.drop(['class'], axis=1)
```

```
classifier = MultinomialNB()
```

```
classifier.fit(x_train, mv_train['class'])
```

```
x_test = mv_test.drop(['class'], axis=1)
classifier.score(x_test, mv_test['class'])
```

```
predict_sentiment = classifier.predict(x_test)
mv_test['prediction'] = predict_sentiment
mv_test
```
